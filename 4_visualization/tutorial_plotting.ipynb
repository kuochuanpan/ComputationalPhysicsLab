```python
import numpy as np
import matplotlib.pyplot as plt
```

### prepare some data for visualization

```python
xx = np.linspace(0,5,100)
y1 = np.sin(xx)
y2 = np.cos(xx)
```

## Line plot

a simplest plot

```python
plt.figure()
plt.plot(xx,y1)
plt.plot(xx,y2)
plt.show()
```

## Scatter Plot

```python
plt.figure(1)
plt.scatter(xx,y1,alpha=0.3)
plt.scatter(xx,y2,alpha=0.3)
```

See more examples:
https://matplotlib.org/stable/gallery/index.html#

# 2D image plots

```python
X, Y = np.meshgrid(np.linspace(-5, 5, 50), np.linspace(-5, 5, 50))
Z  = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
G  = 1./np.sqrt((X-1)**2+(Y)**2) +1./np.sqrt((X+1)**2+(Y)**2) 
```

```python
plt.figure()
plt.imshow(np.log10(G),extent=[-5,5,-5,5],origin="lower") # .T is doing a transpose
plt.xlabel("X")
plt.ylabel("Y")
```
